Find all the hashtags within Tweet
Remove retweet username & tweeted at @username
Remove Links
Remove non-ascii chars
Remove stopwords
Remove email address
Remove digits and special characters

```python
def clean_tweet(tweet):
    hash = re.findall(r"#(\w+)",tweet)
    tweet = re.sub(r"@[A-Za-z0-9-_]+"," ",tweet)
    tweet = re.sub(r"http?://[A-Za-z0-9.]+"," ",tweet)
    cachedStopWords = set(stopwords.words("english"))
    #add custom words
    cachedStopWords.update('and','I','A','http','And','So','arnt','This','When','It','many','Many','so','cant','Yes','yes','No','no','These','these','mailto','regards','ayanna','like','email')
    tweet = " ".join([word for word in tweet.split() if word not in cachedStopWords])
    pattern = re.compile(r"[\w\.]+@[\w\.]+")
    pattern.sub(r" ",tweet)
    tweet = re.sub(r"[^a-zA-Z.!?']", ' ', tweet)

    return tweet
   
```

```python
df['new_tweet'] = df.text.apply(func=clean_tweet)
```
